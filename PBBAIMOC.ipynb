```python
# Python based calculator for basic AI maths operations
```

```python
import numpy as np

# matrix input function
def input_matrix(name):
  print(f"\nEnter values for {name} (e.g. [[1,2],[3,4]])")
  try:
    return np.array(eval(input()))
  except Exception as e:
    print("Invalid input. Try again.")
    return input_matrix(name)

# Vector input function
def input_vector(name):
  print(f"\nEnter values for {name} (e.g. [[1,2],[3,4]])")
  try:
    return np.array(eval(input()))
  except Exception as e:
    print("Invalid input. Try again.")
    return input_vector(name)

def input_value_or_array(name):
  print(f"\nEnter value(s) for {name} (e.g. 5 or [1,2,3])")
  try:
    return np.array(eval(input()))
  except Exception as e:
    print("Invalid input. Try again.")
    return input_value_or_array(name)

def matrix_multiply(A,B):
  try:
    return np.dot(A,B)
  except Exception as e:
        return f"Error: {e}"

def vector_dot(v1,v2):
  try:
    return np.dot(v1,v2)
  except Exception as e:
        return f"Error: {e}"

def sigmoid(x):
  return 1 / (1 + np.exp(-x))

def relu(x):
  return np.maximum(0, x)

def softmax(x):
  e_x = np.exp(x-np.max(x))
  return e_x / e_x.sum(axis=0)

def run_calculator():
  print("Welcome to the AI maths calculator!")

  while True:
    print("\nSelect an operation:")
    print("1. Matrix multiplication")
    print("2. Vector dot product")
    print("3. Sigmoid function")
    print("4. ReLU function")
    print("5. Softmax function")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")
    if choice == '1':
      A = input_matrix("Matrix A")
      B = input_matrix("Matrix B")
      result = matrix_multiply(A, B)
      print(f"\nResult of matrix multiplication:\n{result}")

    elif choice == '2':
      v1 = input_vector("Vector v1")
      v2 = input_vector("Vector v2")
      result = vector_dot(v1, v2)
      print(f"\nResult of vector dot product: {result}")

    elif choice == '3':
      x = input_value_or_array("input for Sigmoid")
      result = sigmoid(x)
      print(f"\nResult of Sigmoid function: {result}")

    elif choice == '4':
      x = input_value_or_array("input for ReLU")
      result = relu(x)
      print(f"\nResult of ReLU function: {result}")

    elif choice == '5':
      x = input_value_or_array("input for Softmax")
      result = softmax(x)
      print(f"\nResult of Softmax function: {result}")

    elif choice == '6':
      print("Exiting calculator. Goodbye!")
      break

    else:
      print("Invalid choice. Please enter a number between 1 and 6.")

run_calculator()
```
